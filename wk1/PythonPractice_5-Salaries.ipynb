## Sanfrancisco Salaries Data Visualization 

```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read the salaries data
data = pd.read_table('./data/Salaries.csv', delimiter=',')
# Print the top 10 rows
data.head(10)
```

```python
##List all Job Titles
data['JobTitle']
```

```python
##List top 10 Job Titles
data['JobTitle'].head(10)
```

```python
# Salary Earners with more than 60000 in Total Pay Benefits
data[data['TotalPayBenefits'] > 60000]
```

```python
# Profession with Highest Pay in San Francisco
data.loc[data['TotalPayBenefits'].argmax()]
```

```python
# Profession with Least Pay in San Francisco
data.loc[data['TotalPayBenefits'].argmin()]
```

```python
# Median TotalPay in San Francisco
data['TotalPayBenefits'].median()
```

```python
# Mean/Average TotalPay in San Francisco
data['TotalPayBenefits'].mean()
```

```python
# Mode TotalPay in San Francisco
data['TotalPayBenefits'].mode()
```

```python
# Top 10 Earners in San Francisco
data.sort_values('TotalPayBenefits',ascending=False).head(10)
```
